```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
eps = 0.7
x = np.linspace(0, 1, num=1000)
ld = 1-eps+eps*(1-x**2)**(1/2)
ld_prime = (-0.7 * x) / (1 - x**2)**(1/2)
```

```python
figsize=(15,7)
titlesize=30
ticksize=25
linesize=5
labelsize=27
legendsize=25
annosize=21
```

```python
plt.figure(figsize=figsize)
plt.title(f"Linear Limb Darkening Function ($\epsilon$={eps})", fontsize=titlesize)
plt.plot([-1.1, 1.1], [1, 1], linewidth=linesize, color='r', linestyle="--")
plt.plot(x, ld, linewidth=linesize, color='g')
plt.plot([0, 0], [0.925, 1.075], linewidth=linesize, color='k')
plt.text(0.025, .86, 'r=0', ha='right', va='bottom', fontsize=annosize, color='k')
plt.plot([0.2, 0.2], [0.925, 1.075], linewidth=linesize, color='k')
plt.text(0.24, .86, 'r=0.2', ha='right', va='bottom', fontsize=annosize, color='k')
plt.xlim(-0.03,1.03)
plt.ylim(0.1,1.1)
plt.xticks(fontsize=ticksize)
plt.yticks(fontsize=ticksize)
plt.xlabel("Normalized Stellar Radius", fontsize=labelsize)
plt.ylabel("Intensity", fontsize=labelsize)
plt.grid(True)
plt.savefig("LD_Funct.pdf", format='pdf')
plt.show()

plt.figure(figsize=figsize)
plt.title(f"Derivative of Limb Darkening Function", fontsize=titlesize)
plt.plot(x, ld_prime, linewidth=linesize, color='g')
plt.xlim(-0.03,1.03)
plt.ylim(-5.1,0.1)
plt.xticks(fontsize=ticksize)
plt.yticks(fontsize=ticksize)
plt.xlabel("Normalized Stellar Radius", fontsize=labelsize)
plt.ylabel("Rate of Change of Intensity", fontsize=labelsize)
plt.grid(True)
#plt.savefig("LD_Funct.pdf", format='pdf')
plt.show()
```
